```python
# read in table with genome size
conc_bin_stats=pd.read_csv('storage/bin_stats.analyze.tsv',header=None,usecols=lambda x: x != 0)
conc_bin_stats.head(5)
```

```python
# split first column that contains bin id 
conc_bin_stats[['Bin_Id',1]]=conc_bin_stats[1].str.split('\t{',expand=True)
```

```python
print(len(conc_bin_stats))
conc_bin_stats.tail(3)
```

```python
# remove bin id col for now, and remove 'healthy_2019_mcav_contigs_cut' bin 
extract_df=conc_bin_stats.iloc[0:189,0:14]
extract_df
```

```python
# in first col of new df: split by ":" and get left side for col names 
new=extract_df.iloc[1].str.split(':', expand=True)
index=new[0]
cols=index.to_list()
```

```python
# loop through and extract the right of : in each row, add to dictionary, readd col names, covnert to df 
d={}
for i in extract_df.index:
    extracted_values = extract_df.iloc[i].str.extract(r'(?<=:)(.*)', expand=True)
    #print(extracted_values)
    list_of_values=extracted_values[0]
    lists=list_of_values.to_list()
    d[i] = lists
d
intermediate=pd.DataFrame.from_dict(d,orient='index')
intermediate.columns=cols
intermediate.tail()
```

```python
#add bin ids back in 
intermediate['Bin_Id']=conc_bin_stats['Bin_Id']
intermediate
```

```python
print(intermediate['Bin_Id'].unique())
print(len(intermediate['Bin_Id'].unique()))
```

```python
concoct_bins.head()
```

```python
# col names became multiIndex, so reformatting
concoct_bins.columns = concoct_bins.columns.get_level_values(0)
```

```python
concoct_bins.Bin_Id.values
```

```python
# converting intermiate bin id values into integers
intermediate["Bin_Id"] = intermediate.Bin_Id.apply(int)
```

```python
#merging dataframes 
concoct_stats_merged=concoct_bins.merge(intermediate,on='Bin_Id')
```

```python
concoct_stats_merged.columns
```

```python
concoct_109=concoct_stats_merged[concoct_stats_merged['Bin_Id'] == 109]
concoct_109
```
