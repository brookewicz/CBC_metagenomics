```python
sample_data=pd.read_csv('Metagenomics_Tracker_Belize.csv', index_col=0)
```

```python
sample_data.head()
```

```python
# make a column for year 

# Extract year from the sample names
def extract_year(sample_name):
    return sample_name.split('_')[0][2:]

# Create a new column 'Year' using the 'apply' function
sample_data['Year'] = sample_data.index.to_series().apply(extract_year)

# Print the updated DataFrame
print(sample_data['Year'])
```

```python
print(sample_data.columns)
```

```python
healthy=sample_data["Health_Status "]=="Healthy"
```

```python
healthy_2019=sample_data[healthy & (sample_data["Year"]=="2019")]
healthy_2019
```

```python
# export list of samples that are healthy, sampled in 2019
healthy_2019_samples = healthy_2019.index.tolist()
```

```python
# Filter for sample names including "MCAV"
healthy_2019_MCAV = [sample_name for sample_name in healthy_2019_samples if "MCAV" in sample_name]

# Print or use the filtered_samples list as needed
print(healthy_2019_MCAV)
```

```python
len(healthy_2019_MCAV)
```

```python
# export list to txt file
file_name = 'healthy_2019_MCAV'

# Write the list to a text file
with open(file_name, 'w') as file:
    for sample_name in healthy_2019_MCAV:
        file.write(f"{sample_name}\n")
```
