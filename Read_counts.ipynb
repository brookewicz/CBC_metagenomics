```python
sample_file='healthy_2019_MCAV_sampleids'
slurm_output='slurm-removal-19358781.out'

# Read sample IDs in order
with open(sample_file, "r") as f:
    sample_ids = [line.strip() for line in f if line.strip()]

# Read SLURM output
with open(slurm_output, "r") as f:
    slurm_data = f.read()

# Split logs to get read counts for each sample
log_blocks = slurm_data.split("reads; of these:")

# first split chunk is SLURM noise, so discard it
log_blocks = log_blocks[1:]

# Prepend "reads; of these:" back to the actual logs
log_blocks = ["reads; of these:" + block for block in log_blocks]

# if # of samples doesn't match log outputs, print and compare
if len(log_blocks) != len(sample_ids):
    print("Warning: Number of logs and samples do not match!")
    print(f"Samples: {len(sample_ids)}")
    print(f"Logs:    {len(log_blocks)}\n")

    # Show side-by-side preview
    for i, (sample, log) in enumerate(zip(sample_ids, log_blocks)):
        print(f"{i}: Sample: {sample} | Log starts with: {log[:50]}")

# Extract unmapped reads 
unmapped_counts = []
for block in log_blocks:
    match = re.search(r"(\d[\d,]*) \([\d.]+%\) aligned 0 times", block)
    #print(match)
    if match:
        seqs = int(match.group(1).replace(",", ""))
        unmapped_counts.append(seqs)
    else:
        unmapped_counts.append(None)

# Extract % mapped 
mapped_percent = []
for block in log_blocks:
    match = re.search(r"([\d.]+)% overall alignment rate", block)
    #print(match)
    if match:
        percent = float(match.group(1))
        mapped_percent.append(percent)
    else:
        mapped_percent.append(None)


# Create dataframe
df = pd.DataFrame({
    "Sample": sample_ids,
    "percent_mapped_host": mapped_percent,
    "Unmapped_Reads": unmapped_counts
})

# divide numbers in half since they are calculating with F & R together 
df['Unmapped_Reads']=df['Unmapped_Reads']/2
# don't know if need to round..

print(df)
```

```python
# append to reads df 
reads = reads.merge(df[['Sample','Unmapped_Reads','percent_mapped_host']], on='Sample', how='left')
```
