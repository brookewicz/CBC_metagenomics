### Search for colonies over time
- trying to find diseased samples from the same healthy mcav colonies over time

```python
cbc_samples=pd.read_csv('CBC_samples.csv', index_col=0)
```

```python
cbc_samples.head()
```

```python
healthy_2019_MCAV
len(healthy_2019_MCAV)
```

```python
# Fill missing values in the 'Tubelabel_species' column with an empty string
cbc_samples['Tubelabel_species'] = cbc_samples['Tubelabel_species'].fillna('')

#search cbc_samples for samples in list
#empty dict
matching_rows = []
# iterate through samples
for sample in healthy_2019_MCAV:
    # find samples in cbc_samples
    matches = cbc_samples[(cbc_samples['Tubelabel_species'].str.contains(sample, case=False)) & (cbc_samples['Sample_type']=="Core_EtOH")]
    # and add to dict
    matching_rows.append(matches)
# convert to df and view
result_df = pd.concat(matching_rows)
result_df

    
```

```python
result_df.shape
#13 rows in result_df matches # of samples in healthy_2019_mcav sample list
```

```python
# if newtagnum is empty, fill with oldtagnum
result_df['NewTagNum'].fillna(result_df['OldTagNum'], inplace=True)
```

```python
# search for associated tag nums in our samples
print(result_df.loc[:, ['NewTagNum', 'Tubelabel_species']])
#save tag numbers as a list
tag_values = result_df['NewTagNum']
print(tag_values)
```

```python
# search for tagnums in sample spreadsheet (to find all samples associated with those colonies)
# Convert tag_values to string
tag_values_str = tag_values.astype(str)
# Search for tag numbers in the 'NewTagNum' column of cbc_samples
matching_tags = cbc_samples[(cbc_samples['NewTagNum'].astype(str).isin(tag_values_str)) & (cbc_samples['Species']=="MCAV") & (cbc_samples['Sample_type']!="Immune")]
# Display the matching rows
matching_tags
```

```python
# Fill missing values in the 'Health status' column with an empty string
matching_tags['Health_status'] = matching_tags['Health_status'].fillna('')

#search for diseased colonies within the matching tags
diseased_matching_tags=matching_tags['Health_status'].str.contains("Diseased", case=False)
diseased_matching_tags.sum()
diseased_mcav_tracking=matching_tags[diseased_matching_tags]
```

```python
# get list of sample ids 
diseased_mcav_list=diseased_mcav_tracking['Tubelabel_species'].tolist()
diseased_mcav_list
```
