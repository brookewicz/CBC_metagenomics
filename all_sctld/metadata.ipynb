```python
# make a column for year 

# Extract year from the sample names
def extract_year(sample_name):
    return sample_name.split('_')[0][2:]

# Create a new column 'Year' using the 'apply' function
samples['Year'] = samples['Sample ID'].apply(extract_year)

# Print the updated DataFrame
print(samples['Year'])
```

```python
# make column for species 
def extract_species(sample_name):
    return sample_name.split('_')[-1]  # Extract last element after splitting by "_"

samples['Species'] = samples['Sample ID'].apply(extract_species)
print(samples['Species'])
```

```python
# make list of all samples 
genohublist24 = samples['Sample ID'].tolist()
genohublist24
```

```python
# replace date error
genohublist24 = [sample.replace("0520220", "052022") for sample in genohublist24]
# fix error 
genohublist24 = [sample.replace("0CBC", "CBC") for sample in genohublist24]
```

```python
# export list to txt file
with open("genohublist24.txt", "w") as f:
    for item in genohublist24:
        f.write(item + "\n")  # Each entry is written on a new line
```
